# Análisis de Medallas Olímpicas

## Cargar los datos

```python
import pandas as pd
df = pd.read_csv("data/medallas.csv")
```

## Exploración Inicial

```python
df.info()
```

## Limpieza de Datos

```python
# Verificamos cuantos datos nulos existen en el dataframe
df.isnull().sum()

# Rellenamos eso datos nulos con "0"
df.fillna(0, inplace=True)
df
```

```python
# Corrección de datos erroneos
df["Oro"] = df["Oro"].astype(int)
df["Plata"] = df["Plata"].astype(int)
df["Bronce"] = df["Bronce"].astype(int)
df
```

## Análisis de Medallas de Oro por País

```python
df.sort_values(by=["Oro"], ascending=False).head(3)
```

## Análisis de Medallas Totales por País

```python
dies_medallas = df[df["Total"] > 10]
dies_medallas.sort_values(by=["Total"], ascending=False)
```
